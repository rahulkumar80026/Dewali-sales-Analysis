```python
# import all libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
# upload the data and apply encoding 
df=pd.read_csv("Diwali Sales Data.csv",encoding='unicode_escape')
df
```

```python
df.shape
```

```python
df.head()
```

```python
df.info()
```

```python
# drop the unrelated data or blank data
df.drop(['Status','unnamed1'],axis=1,inplace=True)
```

```python
# checking the null values
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
#Again check and satisfied that no any null value present in data set
df.isnull().sum()
```

```python
# change the data type 
df['Amount'] = df['Amount'].astype('int')
```

```python
# check again
df.info()
```

```python
df.columns
```

```python
df.describe()
```

```python
# use describe() for specific columns
df[['Age','Orders','Amount']].describe()
```

## Exploratory Data Analysis

```python
# plotting bar chart for gender and it's count
ax = sns.countplot(x = 'Gender',data =df)

# count how much male and female
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gen =df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending =False)
sns.barplot(x="Gender",y='Amount',data =sales_gen)
```

In above graph we can seethat most buyers are females and even the purchasing power of females are greater than males


```python
ax = sns.countplot(data =df, x= 'Age Group',hue='Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Total Amount vs Age Group
sales_age =df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending =True)
#sales_age
sns.barplot(data=sales_age,x='Age Group',y='Amount')
```

From above graph we can see that most of the buyers are of age group between 26-35 females 

## State

```python
# Total number of orders from top 10 states
sales_state =df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders',ascending =False).head(10)
sales_state

sns.set(rc={'figure.figsize':(20,10)})
sns.barplot(data=sales_state,x='State',y='Orders')
```

```python
sales_Order =df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending =False).head(10)
sales_Order

sns.set(rc={'figure.figsize':(20,10)})
sns.barplot(data=sales_Order,x='State',y='Amount')
```

From above graphs we can see that most of the orders and totals amount/sales are from uttar pradesh,Maharashtra and karnataka respectively

## Maritial Status

```python
ax = sns.countplot(data=df,x='Marital_Status')

sns.set(rc={'figure.figsize':(15,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state =df.groupby(['Marital_Status','Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending =False).head(10)
sales_state

sns.barplot(data=sales_state,x='Marital_Status',y='Amount',hue='Gender')

sns.set(rc={'figure.figsize':(15,15)})


```

From above graphs we can see that most of the buyers are married women and they have high purchasing power

## Occupation

```python
ax =sns.countplot(data = df,x='Occupation')
for bars in ax.containers:
    ax.bar_label(bars)
sns.set(rc={'figure.figsize':(30,10)})
```

```python
sales_state =df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending =False).head(10)
sales_state
sns.set(rc={'figure.figsize':(30,15)})
sns.barplot(data = sales_state,x='Occupation',y='Amount')
```

From above graph we can see that most of the buyers are working in IT,Healthcare and Aviation sector

## Product Category

```python
ax = sns.countplot(data=df,x='Product_Category')
sns.set(rc={'figure.figsize':(30,15)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state =df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending =False).head(10)
sales_state
sns.set(rc={'figure.figsize':(30,15)})
sns.barplot(data = sales_state,x='Product_Category',y='Amount')
```

From above graph we can see that most of the sold product are from Food ,Clothing,and Electronic Category

```python
sales_state =df.groupby(['Product_ID'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending =False).head(10)
sales_state

sns.set(rc={'figure.figsize':(30,15)})
sns.barplot(data = sales_state,x='Product_ID',y='Amount')
```

## Conclusion:
Married women age group 26-35 yrs from UP,Maharastra and Karnataka,working in IT,Healthcare,and Aviation are more likely to buy products from Food,Clothing and Electronics Category
